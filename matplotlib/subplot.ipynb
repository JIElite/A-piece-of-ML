## Subplot 

Following is the first method to plot subplot by `plt.subplot`

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
print(x)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# 2-row, 1-column, index: 1
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sine')

# 2-row, 1-column, index: 2
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Can put some parameters in plt.figure(figsize=(x, y)) 
fig = plt.figure(figsize=(12, 4))
plt_sin = fig.add_subplot(1, 2, 1)
plt_sin.plot(x, y_sin)
plt.title("Sine")

plt_cos = fig.add_subplot(1, 2, 2)
plt_cos.plot(x, y_cos)
plt.title("Cosine")

plt.tight_layout()
plt.show()
```
