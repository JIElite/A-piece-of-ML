## Bar plot
reference: https://matplotlib.org/3.1.1/gallery/statistics/barchart_demo.html#sphx-glr-gallery-statistics-barchart-demo-py

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(20)
Y = np.random.randn(20) + 10
print(Y)
```

```python
# Vertical bar plot
plt.bar(X, Y)
plt.show()
```

```python
# Horizontal bar plot
# NOTICE: The value on y-axis 
plt.barh(X, Y)
for i, v in enumerate(Y):
    plt.text(v, i, str('%.2f' % v), color='black', fontweight='bold')
plt.show()
```

```python
X = np.arange(5)
Y = np.random.random(5)
print("X:\n", X)
print("Y:\n", Y)

x_ticks = ['Eli', 'Scarlett', 'Emma', 'Bryant', 'Jason']

plt.bar(X, Y, tick_label=x_ticks)
plt.xlabel('name')
plt.ylabel('value')
plt.show()
```
