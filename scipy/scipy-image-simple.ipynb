reference: https://nthu-datalab.github.io/ml/labs/01_Scientific-Python-101/01_Scientific-Python-101.html

```python
from PIL import Image # PIL: Python Image Library
import scipy
import numpy as np
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
# Please Change the local image here.
image_path = '/home/eli/Pictures/image4.jpg'
```

```python
img_obj = Image.open(image_path)
# img_obj.show()  will pop out a image window.
# we can use imshow to print it in the jupyter notebook
imshow(np.asarray(img_obj))
```

```python
'''
Signature: img_obj.resize(size, resample=0, box=None)
Docstring:
Returns a resized copy of this image.
'''
# **NOT IN-PLACE OPERATION**
img_obj.resize((300, 300))

# Just to a lazy statement to show image
img_obj
```

```python
img = np.asarray(img_obj)
print(img.dtype, img.shape)
print("R:\n", img[:,:,0])
print("G:\n", img[:,:,1])
print("B:\n", img[:,:,2])
```

```python
# tint the image
img_tinted = img * [1, 0.95, 0.9]
print("R:\n", img_tinted[:,:,0])
print("G:\n", img_tinted[:,:,1])
print("B:\n", img_tinted[:,:,2])
```

```python
'''
Signature: Image.fromarray(obj, mode=None)
Docstring:
Creates an image memory from an object exporting the array interface
(using the buffer protocol).
'''
# mode: 
img_tinted = Image.fromarray(img_tinted, "RGB")
img_tinted = img_tinted.resize((300, 300))
img_tinted.save('tinted_image.jpg')
img_tinted
```

## Distance between points

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform, sqeuclidean
# pdist means: pair-wise distance

x = np.array([
    [0, 1],
    [1, 0],
    [2, 0]
])

pd = pdist(x, 'euclidean')
print(pd)

# d[i, j] is the Euclidean distance between x[i,:], x[j:,] ()
# and d is as following:
d = squareform(pd)
print(d)
```
